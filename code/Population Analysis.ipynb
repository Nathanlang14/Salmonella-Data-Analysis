```python
import pandas as pd
```

```python
data = pd.read_csv('../csvs/df_state.csv')
```

```python
data.head()
```

```python
# drops California because its an outlier (although still on relative line)
data = data.drop(4)
```

```python
%matplotlib inline 

import seaborn as sns
sns.set(color_codes = True)
sns.regplot(y=data['Hospitalizations'], x = data['Illnesses']).set_title('Hospitalizations vs Illnesses')
```

```python
data['Percent'] = 100*data.Illnesses/data.Pop
data['Death_per'] = 100*data.Deaths/data.Hospitalizations
```

```python
%matplotlib inline
data.plot('State', 'Death_per', kind = 'bar', title = 'Percent of Population Infected (2009-2016)')
```
